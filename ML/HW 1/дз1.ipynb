## №1
Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы

1.1

```python
import numpy as np 
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])

def loops1(X):
    res1 = 1
    n = min(len(X), len(X[0]))
    for i in range (n):
        if X[i][i] !=0:
            res1*=X[i][i]
    return res1
print(f'Ответ решения циклами: {loops1(X)}')
```

1.2

```python
def np_only1(X):
    return np.prod((np.diag(X)[np.diag(X)!=0]))
print(f'Ответ решения только при помощи Numpy: {np_only1(X)}')
```

1.3

```python
def mixed1(X):
    res3 = 1
    for i in (np.nonzero(np.diag(X))[0]):
        res3*=X[i][i]
    return res3
print(f'Ответ смешанного решения: {mixed1(X)}')
```

```python
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])

x=np.sort(x)
y=np.sort(y)

def loops2(x,y):
    flag = True
    for i in range(len(x)):
        if (x[i]!=y[i]):
            flag = False
            break
    return flag
        
print(f'Решение циклами: {loops2(x,y)}')
```

## №3
Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.

3.1


```python
x3 = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

def loops3(x):
    max = 0
    for i in range (len(x)):
        if x[i]==0 and i<(len(x)-1):
            if x[i+1]>max:
                max = x[i+1]
    return max
print(f'Решение циклами: {loops3(x3)}')
```

3.2

```python
def np_only3(x):
    ind = np.where(x==0)[0]
    ind1=(np.where(ind<(len(x)-1), ind, 0))
    return max(x[ind1[np.nonzero(ind1)[0]]+1])  
print(f'Решение numpy: {np_only3(x3)}')

    
```

3.3

```python
def mixed3(x):
    ind = np.where(x==0)[0]
    for i in range(1,len(x)):
        if i-1 not in ind:
            x[i]=0
    return np.max(x[1:])
print(f'Смешанное решение: {mixed3(x3)}')
```

## №4
Дан трёхмерный массив, содержащий изображение, размера (height, width, numChannels), а также вектор длины numChannels. Сложить каналы изображения с указанными весами, и вернуть результат в виде матрицы размера (height, width). Считать реальное изображение можно при помощи функции scipy.misc.imread (если изображение не в формате png, установите пакет pillow: conda install pillow). Преобразуйте цветное изображение в оттенки серого, использовав коэффициенты np.array([0.299, 0.587, 0.114]).

4.1

```python
import matplotlib.pyplot as plt
img = plt.imread('picture1.jpg')

coef = np.array([0.299, 0.587, 0.114])

def loops4(img, coef):
    h = len(img)
    w = len(img[0])
    res = []
    for i in range(h):
        curr =[]
        for j in range (w):
            sum = 0
            for k in range (len(coef)):
                sum += img[i][j][k]*coef[k]
            curr.append(sum)
        res.append(curr)
    return res



```

4.2

```python
def np_only4(img, coef):
    return np.sum(img*coef, axis = -1)

```

4.3

```python
def mixed4(img, coef):
    res = np.zeros((len(img), len(img[0])))
    for i in range (len(img)):
        for j in range (len(img[0])):
            res[i][j]=(np.sum(coef*img[i][j]))
    return res

```

```python

plt.figure(figsize=(15,8))
plt.subplot(221)
plt.imshow(img)
plt.title("Оригинал")
plt.subplot(222)
plt.imshow(np_only4(img, coef), cmap = 'gray')
plt.title("numpy")
plt.subplot(223)
plt.imshow(loops4(img, coef), cmap = 'gray')
plt.title("Циклы")
plt.subplot(224)
plt.imshow(mixed4(img, coef), cmap = 'gray')
plt.title("Смешанный")
plt.show()
```

```python
%timeit loops4(img, coef)
```

```python
%timeit np_only4(img,coef)
```

```python
%timeit mixed4(img, coef)
```

## Вывод
Как мы видим, быстрее всего работает алгоритм использующий только numpy

## №5 
Реализовать кодирование длин серий (Run-length encoding). Дан вектор x. Необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.

5.1

```python
x = np.array([2, 2, 2, 3, 3, 3, 5, 2, 2])

def loops5(x):
    nums = []
    count = []
    c = 1
    for i in range(1,len(x)):
        if x[i-1]==x[i]:
            c+=1
        else:
            nums.append(x[i-1])
            count.append(c)
            c=1
    nums.append(x[i])
    count.append(c)
    return (nums, count)

print(loops5(x)) 
```

5.2

```python
def np_only5(x):
    y = np.hstack((np.ones(1), x[:- 1]))
    first_positions = x != y
    first_positions[0] = True
    indexes_1 = np.arange(np.size(x))[first_positions]
    indexes_2 = np.hstack((indexes_1[1:], np.array([np.size(x)])))
    return x[first_positions], indexes_2 - indexes_1
   


print(np_only5(x))
```

```python
def mixed5(x):
    u, indices = np.unique(x, return_inverse=True)
    nums =[]
    count = []
    c=1
    nums.append(u[indices[0]])
    for i in range (1,len(indices)):
        if indices[i]==indices[i-1]:
            c+=1
        else:
            nums.append(u[indices[i]])
            count.append(c)
            c=1
    count.append(c)
    return nums,count
            
print(mixed5(x))
```
